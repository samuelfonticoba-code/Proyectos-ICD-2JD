## ***¿Qué es ese azúcar en sangre?:***
#### Para explicarlo de forma sencilla, el azúcar en sangre o glucosa es el combustible principal que usan tus células para llegar a moverse, pensar y vivir. Obtenemos esto mediante el consumo de carbohidratos (frutas, dulces, etc.). El cuerpo los descompone en glucosa y la lanza a su torrente sanguíneo. Un nivel demasiado alto (hiperglucemia) es dañino y se asocia con la diabetes.
#### Pero, ¿qué es ese "azúcar"? Gracias a comer carbohidratos, el cuerpo los descompone en glucosa y la lanza a su torrente sanguíneo.
#### hora, para lograr regular esta y mantener al cuerpo en su punto correcto, nuestro organismo utiliza algo llamado insulina, de vital importancia, generado por nuestro páncreas. Al igual que un panadero sabe cómo, cuándo y mediante qué hacer una buena oblea, nuestro cuerpo funciona así. La insulina disminuye la glucemia, actúa como señal para que las células capten y utilicen la glucosa del torrente sanguíneo, o la almacenen como glucógeno (en hígado, músculo o grasa). Existe otra variante de la que hablamos: glucemia baja o alta. Para el primer caso (hipoglucemia), para esto nuestro páncreas aumenta la glucemia, donde señaliza al hígado para que libere glucosa de sus reservas o la sintetice cuando el nivel de azúcar es bajo.
#### Y ciertamente puede sonar como algo tratable, pero no es tan sencillo como quizás parece; pues, al igual que un motor de automóvil deja de funcionar, puede causar grandes consecuencias. ## **¿Los alimentos solo se componen de esto?**

#### No, para que una dieta pueda tener sentido, mínimamente tendría que estar balanceada y este es el caso; por lo tanto, no me baso solo en esto para determinarla. Según investigo, existen tres macronutrientes principales que poseen los alimentos y que componen, en mayor medida, nuestro organismo para realizar sus funciones básicas:

* ####  ***```Proteinas```***:Función: Estructural y enzimática. Son las responsables de la síntesis de tejidos (músculo, piel, órganos), la respuesta inmunitaria y la creación de hormonas.
* ####  ***```Lipidos```***:Función: Reserva energética, aislamiento térmico y protección de órganos. Son fundamentales para la estructura de las membranas celulares y la absorción de vitaminas
* ####  ***```Carbohidratos```***:Suministro energético inmediato. Son la fuente de combustible preferida por el Sistema Nervioso Central y los eritrocitos (glóbulos rojos). 

#### Ahora, para explicarlos de manera más sencilla: las proteínas son las encargadas de las funciones estructurales en nuestro cuerpo; se encargan de la reparación y construcción de tejidos, así como del aumento de la masa muscular o reconstrucción de tejido adiposo (piel). En cierto sentido, es como tu albañil: se encarga de todo tipo de estructura, pero si no lo mantienes en tu obra, no tiene sentido seguir; se derrumba {como La Habana ;)}.

#### Los carbohidratos tienen un papel energético. ¿Para qué sirven? Son la energía rápida; es lo que usa tu cerebro para pensar y tus piernas para caminar. En sí mismos, tienen un papel como combustible (que anda en falta). Nuestro cuerpo es como un motor que necesita un medio de obtención de energía para hacer girar sus rotores, y nuestra gasolina, en este caso, son nuestros mayores contenedores de carbohidratos en la dieta.

####  ***A medida que envejecemos, el cuerpo pierde su capacidad de resiliencia metabólica.***
#### ***El Tiempo se te termina y tu no lo ves pasar***

#### Los órganos de nuestro cuerpo, al igual que cualquier maquinaria de precisión, sufren un desgaste inevitable con el paso del tiempo. Sus funciones no son eternas, basta con observar a un deportista de élite: un futbolista, tras alcanzar cierta edad, debe retirarse. No es solo una cuestión de contrato, sino de rendimiento. En la vejez, el margen para los errores dietéticos se cierra, y lo que antes era un esfuerzo rutinario para el páncreas, se convierte en una batalla perdida.
